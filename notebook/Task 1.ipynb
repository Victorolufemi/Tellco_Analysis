### 1.2

##### Performing Data Cleaning

```python
data = eda.fix_cat_cols(data, cat_col)
```

```python
data = eda.fix_num_cols(data, num_col)
```

```python
data.info()
```

```python
outliers = ['Avg RTT DL (ms)', 'Avg RTT UL (ms)','TCP DL Retrans. Vol (MB)','TCP UL Retrans. Vol (MB)','Avg Bearer TP DL (kbps)','Avg Bearer TP UL (kbps)']
```

```python
data = eda.fix_outlier(data, outliers)
```

###### 	Describe all relevant variables and associated data types  

```python
df = pd.DataFrame()
df['Dur. (s)'] = data['Dur. (ms)']/1000
df['Total UL (MB)'] = data['Total UL (MB)']
df['Total DL (MB)'] = data['Total DL (MB)']
df['Google_total (MB)'] = data['Google DL (MB)'] + data['Google UL (MB)']
df['Social_total (MB)'] = data['Social Media DL (MB)'] + data['Social Media UL (MB)']
df['Email_total (MB)'] = data['Email DL (MB)'] + data['Email UL (MB)']
df['Youtube_total (MB)'] = data['Youtube DL (MB)'] + data['Youtube UL (MB)']
df['Netflix_total (MB)'] = data['Netflix DL (MB)'] + data['Netflix UL (MB)']
df['Gaming_total (MB)'] = data['Gaming DL (MB)'] + data['Gaming UL (MB)']
df['Other_total (MB)'] = data['Other DL (MB)'] + data['Other UL (MB)']
df['Total_data (MB)'] = data['Total_Data_Volume(MB)']
df['MSISDN/Number'] = data['MSISDN/Number']
df.head()
```

```python
df.info()
```
